```python
import pandas as pd
import numpy as np

# from bokeh.io import output_file
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
output_notebook()

#load in the metadata file from the shared google drive
original_meta_data = pd.read_csv('C:/Users/Caleb/Desktop/Poolla Lab/Data/metadata.csv') #need to change this to match your google drive path, but should always end with .../CPS Project/Raw Data/metadata.csv

original_meta_data.head(10)

```

```python
 #this function will see if any columns are completely empty and create a new df as its output with those columns removed
 
 def check_all_nan_and_delete(df):
  total_non_nan=np.sum(df.notnull())==0
  new_df=df.loc[:,~total_non_nan]
  if sum(total_non_nan)==0:
    print("No Columns Deleted. All Columns have at least one entry")
    return df
  else:
    deleted_cols=df.loc[:,total_non_nan].columns
    print("column(s) deleted:",deleted_cols)
    return new_df

  


meta_data=check_all_nan_and_delete(original_meta_data)

#lets work with this df from now on even though it is the same as meta_data
```

```python
#lets turn all the boolean columns in the meta_data into proper boolean datatype
def is_yes(series):
  return series=='yes'

just_strings=meta_data.select_dtypes(include=object).apply(is_yes,1)
dont_boolean=just_strings.apply(lambda x: sum(x))
dont_boolean=dont_boolean[dont_boolean==0]

only_booleans=just_strings.drop(columns=dont_boolean.index) 

#now lets integrate our new boolean table into the meta_data

meta_data[only_booleans.columns] = only_booleans
meta_data.head(5)



```

```python
#now lets see how many houses have EV and PV
#first group by state

meta_data.groupby('state')[['dataid']].agg(len).rename(columns={'dataid':'count'})

```

```python
# Now lets search by state for homes with both EV and PV

EV_PV_index= meta_data['pv'] & meta_data['car1']

EV_PV_homes=meta_data.loc[EV_PV_index,:]

print("Total count of households with PV and an EV by state data:\n")
EV_PV_homes.groupby('state').agg(len).iloc[:,1]


```

```python
#now lets grab the household id numbers "dataid" of houses with PV and EV so we can compare with the household energy data we have at different intervals

dataids_PV_EV_homes = list(EV_PV_homes['dataid'])
dataids_PV_EV_homes
```

PART 2: Find our PV & EV homes with the energy use datasheets

```python
#load in our data sets

original_TX_15min = pd.read_csv("C:/Users/Caleb/Desktop/Poolla Lab/Data/15minute_data_austin.csv.gz",compression='gzip')
original_CA_15min = pd.read_csv("C:/Users/Caleb/Desktop/Poolla Lab/Data/15minute_data_california.csv.gz",compression='gzip')
original_NY_15min = pd.read_csv("C:/Users/Caleb/Desktop/Poolla Lab/Data/15minute_data_newyork.csv.gz",compression='gzip')

#this is all the 15 min interval data available on PECAN STREET as of 3/26/2020

```

```python
#add a state column to all thagg(ese and merge into one table to view
def give_state_col(df,state_str):
  df["state"]=state_str
  return df


edited_15min_CA=give_state_col(original_CA_15min,"CA")
edited_15min_NY=give_state_col(original_NY_15min,"NY")
edited_15min_TX=give_state_col(original_TX_15min,"TX")


#lets join all table into a master table for 15 min intervals
combined_15min=edited_15min_CA.append(edited_15min_NY,ignore_index=True)
combined_15min=combined_15min.append(edited_15min_TX,ignore_index=True)



```

```python
#lets see how many homes are in this combined dataset and how many points at each home were taken total
home_overview=combined_15min.groupby("dataid").agg(len)
home_overview[["local_15min"]].rename(columns={"local_15min":"rows_data_available"})
```

```python
#lets make a method to take in a Pecan St energy usage data file and give us only the data for homes with an EV and PV charger in a new df


def get_pv_ev_homes(raw_df,dataids_list):
  new_df=raw_df.loc[raw_df['dataid'].isin(dataids_list)]
  new_df

  return new_df
```

```python
#now lets use the method above to confirm that the list of EV and PV homes from the meta data matches the combined 15 min interval dataframe


#get a df of EV and PV homes from the combined dataset using list created from meta data
combined_EV_PV=get_pv_ev_homes(combined_15min,list(EV_PV_homes["dataid"]))

#create a subset of this data that actually has PV and EV data
sorting_df=combined_EV_PV.groupby("dataid").agg(sum)
sorting_df= sorting_df.loc[(sorting_df["car1"]!=0) & (sorting_df["solar"]!=0),:]
final_PV_EV_dataids=list(sorting_df[["car1","solar"]].index)

#do the same as above but using the combined 15 min interval dataset and not going off of what the meta data says should have EV and PV
sorting_df2=combined_15min.groupby("dataid").agg(sum)
sorting_df2= sorting_df2.loc[(sorting_df2["car1"]!=0) & (sorting_df2["solar"]!=0),:]
final_PV_EV_dataids2=list(sorting_df2[["car1","solar"]].index)


#compare the two approaches above
final_PV_EV_dataids==final_PV_EV_dataids2 #false
[final_PV_EV_dataids,final_PV_EV_dataids2]


#as of now, it looks like the metadata is not correct for CA and NY because there are homes with 0 net pv gen and 0 net car1 gen in both states (TX is fine).
#also there is 1 home that we discover fits our EV and PV criteria that is not shown in the metadata
#no CA houses end up meeting our criteria
```

```python
all_states_15min_EV_PV_homes_only=get_pv_ev_homes(combined_15min,final_PV_EV_dataids2)
print("Homes with EV and PV 15min interval data available: \n")
all_states_15min_EV_PV_homes_only.groupby(["state","dataid"]).agg(len)[["air1"]].rename(columns={"air1":"rows_data_available"})
```
